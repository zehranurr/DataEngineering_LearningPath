**Local Language Model (LLM) Integration**

**ChatOllama:**
You integrate a fully local AI model into your project that runs entirely on your own machine, without sending any data externally (e.g., to OpenAI servers).

**model="qwen3:4b-instruct-2507-q4_K_M":**
Specifies the exact model to be run via Ollama. This is a relatively lightweight Qwen model optimized for instruction-following tasks.

**temperature=0:**
This setting is critical. It completely disables the model’s creativity. In RAG systems, setting the temperature to 0 helps prevent hallucinations and ensures that the model generates precise answers strictly based on the provided documents.


Text Embedding Models

This section is the core component that enables hybrid search. You convert text into numerical representations (vectors) that machines can understand.

dense_embeddings (HuggingFaceEmbeddings):
This acts as your semantic search engine. The all-mpnet-base-v2 model captures the meaning and context of sentences. For example, when a user searches for “automobile,” it can also retrieve documents containing related terms like “car” or “vehicle.”

sparse_embeddings (FastEmbedSparse):
This serves as your keyword-based search engine. It uses the Qdrant/BM25 algorithm. Similar to traditional search engines, it matches exact terms from the user’s query (such as proper nouns or product codes) directly with the documents.
